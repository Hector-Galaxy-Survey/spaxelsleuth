## Shortlist of high-S/N galaxies
A subset of galaxies were selected for re-fitted with LZIFU. 
Only high-S/N galaxies were chosen to maximise the reliability of the stellar and emission line fitting. 
For each galaxy, we computed the median blue continuum S/N in each spaxel, where the median was computed over the entire wavelength range in the cube.
We then computed the median of these values over the entire FoV.
